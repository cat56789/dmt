```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
INPUT_CSV = "/content/kmeans_dataset.csv"
OUTPUT_CSV = "kmeans_dataset_labeled.csv"
PLOT_PATH = "kmeans_clusters.png"
K = 2
RANDOM_STATE = 42
```

```python
df = pd.read_csv(INPUT_CSV)

if not {"x", "y"}.issubset(df.columns):
    raise ValueError("CSV must contain columns 'x' and 'y'")

X = df[["x", "y"]].values
```

```python
kmeans = KMeans(n_clusters=K, random_state=RANDOM_STATE, n_init=10)
kmeans.fit(X)
labels = kmeans.labels_
centroids = kmeans.cluster_centers_
```

```python
df["cluster"] = labels
df.to_csv(OUTPUT_CSV, index=False)
print(f"Labeled CSV saved to: {OUTPUT_CSV}")
```

```python
plt.figure(figsize=(8, 6))

for cluster_id in range(K):
    mask = labels == cluster_id
    plt.scatter(X[mask, 0], X[mask, 1], s=60, label=f"Cluster {cluster_id}", alpha=0.8)
```

```python
plt.figure(figsize=(8, 6))

for cluster_id in range(K):
    mask = labels == cluster_id
    plt.scatter(X[mask, 0], X[mask, 1], s=60, label=f"Cluster {cluster_id}", alpha=0.8)

plt.scatter(centroids[:, 0], centroids[:, 1],
            marker="X", s=200, linewidths=1, edgecolor="k", label="Centroids")

plt.title(f"K-Means Clustering (k={K})")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig(PLOT_PATH, dpi=150)
plt.show()
print(f"Cluster plot saved to: {PLOT_PATH}")
```
