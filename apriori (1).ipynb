```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
df = pd.read_csv("/content/bread basket.csv")
df.head()
```

```python
basket = (df.groupby(['Transaction', 'Item'])['Item'].count().unstack().reset_index().fillna(0).set_index('Transaction'))
```

```python
basket = basket.astype(bool).astype(int)
print(basket.head())
```

```python
frequent_itemsets = apriori(basket, min_support=0.02, use_colnames=True)
```

```python
print("\nFrequent itemsets found:")
print(frequent_itemsets.sort_values('support', ascending=False).head(20))
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1.0)
```

```python

print("\nAssociation rules:")
print(rules.sort_values(['lift', 'confidence'], ascending=False).head(20))

frequent_itemsets.to_csv("apriori_frequent_itemsets.csv", index=False)
rules.to_csv("apriori_association_rules.csv", index=False)

print("\nSaved 'apriori_frequent_itemsets.csv' and 'apriori_association_rules.csv'")
```
